```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd

results = pd.read_csv("results-basic.csv").sort_values("probability", ascending=False)

# slider to control how many results to show
slider = widgets.IntSlider(value=20, min=5, max=len(results), step=5, description="Top N:")

def show_results(n):
    top_n = results[["Name", "probability"]].head(n).reset_index(drop=True)
    top_n.index += 1
    top_n.columns = ["Building Name", "Designation Likelihood"]
    display(top_n.style
        .background_gradient(subset=["Designation Likelihood"], cmap="YlOrRd")
        .format({"Designation Likelihood": "{:.1%}"})
        .set_properties(**{"text-align": "left"})
        .set_table_styles([{"selector": "th", "props": [("font-weight", "bold")]}])
    )

```

```python
widgets.interact(show_results, n=slider)
```
